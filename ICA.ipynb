```python
import SimpleITK as sitk
import nibabel as nib
import matplotlib.pyplot as plt
from nilearn import image as nii
from nilearn import plotting

from nilearn.decomposition import CanICA
```

```python
"""We create ICA in order to see independent components of brain, 
here we see Mask_strartegy as template which uses grey compound of brain to provide independent components """
path = r'C:\Users\sireeshan\Downloads\ML Neuroscience Task - Helper Data\ML Neuroscience Task - Helper Data\Preprocessed rs-fMRI scan\Filtered_4DVolume.nii'
#image = sitk.ReadImage(path)
canica = CanICA(n_components=20,
                memory="nilearn_cache", memory_level=2,
                verbose=10,
                mask_strategy='template',
                random_state=0)
canica.fit(path)
```

```python

canica_components_img = canica.components_img_
ica_path = r'C:\Users\sireeshan\Downloads\ML Neuroscience Task - Helper Data\ML Neuroscience Task - Helper Data\Preprocessed rs-fMRI scan\canica_resting_state.nii.gz'
canica_components_img.to_filename(ica_path)
```

```python

from nilearn.plotting import plot_prob_atlas

# Plot all ICA components together
plot_prob_atlas(canica_components_img, title='All ICA components')

```

```python

from nilearn.image import iter_img
from nilearn.plotting import plot_stat_map, show
# displaying all components
for i, cur_img in enumerate(iter_img(canica_components_img)):
    plot_stat_map(cur_img, display_mode="z", title="IC %d" % i,
                  cut_coords=1, colorbar=False)
```
